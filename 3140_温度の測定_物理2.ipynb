最後の目盛の温度は, カ氏温度 ℉ が利用されています.

x ℉ から y ℃ への換算式は

> $ x = \dfrac{9}{5} y + 32 $

> $ y = \dfrac{5}{9}\left(x-32\right)$

です. よって図の目盛りは およそ 23-34℃ と換算できます.
