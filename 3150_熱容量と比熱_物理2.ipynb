- 物体が吸収/放出する熱量にその物体の温度差 $\Delta T = T -T_0$ が比例する.
\begin{align*}
Q \propto \Delta T 
\end{align*}
- 注目する物体(水, 油, 鉄, 銅等)の塊によって，温度差 $\Delta T$ が異なる.
\begin{align*}
Q = C \cdot\Delta T 
\end{align*}
ここで現れた $C$ を **熱容量** (単位 J/K, cal/K) という.
- 熱容量を, 注目する物体の単位質量 1g あたりに注目した値を **比熱** という.
  - 物体の塊の重さを g 質量単位で $m$, その熱容量を $C$ なら 比熱は $c=\dfrac{C}{m}$
    - 変形すれば $C=m\cdot c$
  - 比熱の単位は (単位 J/g.K, cal/g.K) 
- 熱容量 $C$, 比熱 $c$ の単位に現れる K は, 温度差 $\Delta T$ なので ℃ 変換できます

これらをまとめると, 顕熱[物体の温度を明示的に変化できる熱量]では次の関係式が成り立ちます.
\begin{align*}
Q &= C \Delta T&\\
&= c m \Delta T
\end{align*}
この式から
- 比熱 $c = \dfrac{Q}{m\Delta T}$
- 温度変化 $\Delta T = \dfrac{Q}{cm}$
  - $Q/m$ が一定で, 比熱 $c$ が大きければ, $\Delta T$ の変化は小さく,
  - $Q/m$ が一定で, 比熱 $c$ が小さければ, $\Delta T$ の変化は大きくなります.

異種物体間の熱量保存則は,各物体が吸収した熱量 $Q_{\rm in}$ と
放出した熱量 $Q_{\rm out}$ を計算することで得られます.

1. まずは異種間の物体の初期温度と熱平衡に至った後の温度関係に注意して下さい. 上の図では
> $T_1 < T_3 < T_2$

2. 低温の物体(1)が吸収した熱量 $Q_in$ は, 顕熱を温度変化で表すと
> $Q_{\rm in} = m_1 c_1 \left(T_3 - T_1\right)$

3. 高温の物体(2)が放出した熱量 $Q_out$ は, 顕熱を温度変化で表すと
> $Q_{\rm out} = m_2 c_2 \left(T_2 - T_3\right)$

4. この接触では**熱量保存則** が成り立ち, 熱量は物体間のみで行われたとすると
> \begin{align*}
Q_{\rm out} &= Q_{\rm out} &\\
m_2 c_2 \left(T_2 - T_3\right)
&=
m_1 c_1 \left(T_3 - T_1\right)&
\end{align*}

このような関係を利用して, 既知の物体の比熱 $c_0$ を利用して未知の物体の比熱 $c$ や熱容量 $C=m\cdot c$ を求めることができます.

- 測りたい試料の情報: 質量 $m$, 比熱(未知) $c$, その初期状態の温度(測定可能) $T$
- 熱量計内  
  - 水の情報: 質量 $m_{\rm w}$, 比熱(既知) $c_{\rm w}$, 初期状態の温度(測定可能) $T_0$
  - 容器とかき混ぜ棒の情報: 質量 $m_0$, 比熱(既知) $c_0$, 初期状態の温度(測定可能) $T_0$

測りたい試料を初期温度$T$に温めておいて, 熱量計の中に投入し, ある程度時間が経てば
- 試料も熱量計も熱平衡温度 $T'$ に至る
- 温度の大小関係: $T_0 < T' < T$



これらを **熱量保存則** を用いて次のように整理できます.
1. 熱量計全体が受け取る熱量 $Q_{\rm in}$ は
  - 水が吸収する熱量: $m_{\rm w} c_{\rm w} (T'-T_0)$
  - 容器とかき混ぜ棒が吸収する熱量: $m_0 c_0 (T'-T_0)$
> $\qquad Q_{\rm in} = (m_{\rm w} c_{\rm w}+m_0 c_0)(T'-T_0)$

2. 試料が放出する熱量 $Q_{\rm out}$ は
> $\qquad Q_{\rm out} = m c (T-T')$

3. 熱量計内で熱の出入りが完結していると考えれば, 熱量保存則を適用し
\begin{align*}
Q_{\rm in} &= Q_{\rm out}&\\
(m_{\rm w} c_{\rm w}+m_0 c_0)(T'-T_0) &= m c (T-T')&\\
\end{align*}
これを変形すると
>>\begin{align*}
c &= \dfrac{(m_{\rm w} c_{\rm w}+m_0 c_0)}{m}\cdot
\dfrac{T'-T_0}{T-T'}&
\end{align*}

と試料の比熱を測定することができます.




```
#@title 計算その1
(cw,mw,T0)=(4.2,80,15)
(T1,m1,T1d)=(43,70,27)
(T2,m2,T2d)=(97,45,29)
C=cw*(m1*(T1-T1d)/(T1d-T0) -mw)
print(C)
```

```
#@title 計算その2
mw2=mw+m1
c2=(mw2*cw+C)/m2*(T2d-T1d)/(T2-T2d)
print(c2)
```
