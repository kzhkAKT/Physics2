#### 潜熱

Black(1762) が Latenet Heat と命名
> 二つの相を転移する間に **潜** む **熱**

>> $Q_{\rm L} = m \cdot L$

ここで $m$ は 単位質量 1g, $L$ は潜熱(融解熱, 気化熱の総称)

- 注目する物質の量(質量)に依存
- 物質の種類(水,金属等)によって異なる
- 固相から液相へ転移する温度が *融解点* で, そのとき単位質量(1g)あたり *吸収する熱量* を **融解熱** 
  - 液相から固相へ転移する温度は同じ温度だが *凝固点* といい,
   そのとき *放出する熱量* も融解熱と同量だが **凝固熱** 
- 液相から気相へ転移する温度が *沸点* で,
そのとき単位質量(1g)あたり*吸収する熱量* を **気化熱**/蒸発熱
  - 気相から液相へ転移する温度は同じ温度だが *液化点*/凝集点 といい,
   そのとき *放出する熱量* も気化熱と同量だが **液化熱**/凝縮熱

```
#@title 例題3.3 の解答
import sympy
T = sympy.symbols('Th')
(C,Th,mw)=(84,60,180)
(mi,Ti)=(50,0)
(cw,Tw, Liw)=(4.2,0,3.3*10**2)
Qin=mi*Liw + cw*mi*(T-Tw)
Qout=C*(Th-T) + cw*mw*(Th-T)
ansT=sympy.solve(Qin-Qout, (T))
print("{:.3f}".format(ansT[0]))
```
