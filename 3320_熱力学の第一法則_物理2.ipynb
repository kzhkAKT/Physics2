# 熱力学の第一法則

## 復習: 内部エネルギー $U$

注目する物体の内部エネルギー $U$ は, 
>> $U$ = **全気体分子の運動エネルギー** +  **気体分子間の相互作用**ポテンシャル

でした.

理想気体では、相互作用を無視するので
>> $U$ = **全気体分子の運動エネルギー**

となります。

## 内部エネルギーを変化させる方法: 熱力学の第一法則

\begin{align*}
\Delta U &= Q_{\rm in} + W_{\rm in}&
\end{align*}
注目する系(ピストンに閉じ込められた気体等)が  
エネルギーを受け取る方向を **in** とし **正(+)** とすると,   
気体が外部にする仕事 $W_{\rm out}$ と $W_{\rm in}$ の関係は
\begin{align*}
\Delta U &= Q_{\rm in} + W_{\rm in}&\\
&= Q_{\rm in} - W_{\rm out}&
\end{align*}

講義では系に加わる熱量や仕事も変化量として
\begin{align*}
\Delta U & = \Delta Q_{\rm in} + \Delta W_{\rm in}&\\
& = \Delta Q_{\rm in} - \Delta W_{\rm out}&\\
\end{align*}
と記述します.
